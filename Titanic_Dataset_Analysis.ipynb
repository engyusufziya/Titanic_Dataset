```python
sns.barplot(data=df, x='Embarked', y='Survived')
plt.title("Sıralı Göre Hayatta Kalma Oranı")
plt.show()
```

*   Yolcuların bindikleri limana (S, C, Q) göre ortalama hayatta kalma oranlarını görüyoruz. En yüksek hayatta kalma oranı Cherbourg (C) limanından binen yolcularda, en düşük oran ise Southampton (S) limanından binenlerde gözleniyor.
*   Cherbourg’tan binenlerin arasında daha fazla 1. sınıf yolcu olduğu düşünülüyor; bu da hayatta kalma oranını artırmış olabilir. Southampton ise geminin ilk kalkış noktasıydı ve en fazla yolcunun bindiği limandı; bu nedenle yoğunluk ve sınıfsal dağılım gibi faktörler hayatta kalma oranını düşürmüş olabilir.
